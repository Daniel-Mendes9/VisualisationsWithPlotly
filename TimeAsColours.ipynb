```python
import math
import plotly.express as px
import pandas as pd

pontos = []
amplitude = math.sqrt(1 + 1)  

for t in range(0, 201):
    t = t / 10  
    theta = math.acos(1 / amplitude) * math.cos(t / math.sqrt(9.8))
    x = amplitude * math.sin(theta)
    y = amplitude * math.sin(theta)
    z = -amplitude * math.cos(theta)
    pontos.append([x, y, z])

# Crie um DataFrame a partir dos pontos
df = pd.DataFrame(pontos, columns=["Distancia x", "Distancia y", "Distancia z"])

fig = px.line_3d(df, x="Distancia x", y="Distancia y", z="Distancia z")
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np
import math

amplitude = math.sqrt(1 + 1)  

t = np.linspace(0, 30, 100)
theta = [math.acos(1 / amplitude) * math.cos(ti / math.sqrt(9.8)) for ti in t]
x = amplitude * np.sin(theta)
y = np.zeros(len(t))
z = -amplitude * np.cos(theta)+2

fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=5,
        color=z,               
        colorscale='Viridis',   
        opacity=0.8
    )
)])


fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np
import math

amplitude = math.sqrt(1 + 1)

t = np.linspace(0, 30, 100)
theta = [math.acos(1 / amplitude) * math.cos(ti / math.sqrt(9.8)) for ti in t]

#Primeira Srie de dados
x1 = amplitude * np.sin(theta)
y1 = amplitude * np.sin(theta)
z1 = -amplitude * np.cos(theta) + 2

# Segunda srie de dados
x2 = amplitude * np.sin(theta)
y2 = np.zeros(len(t))
z2 = -amplitude * np.cos(theta) + 2  

# Terceira srie de dados
x3 = amplitude * np.sin(theta)
y3 = -amplitude * np.sin(theta)# Diferente de z1
z3 = -amplitude * np.cos(theta) + 2 

fig = go.Figure()

fig.add_trace(go.Scatter3d(
    x=x1,
    y=y1,
    z=z1,
    mode='lines+markers',
    marker=dict(
        size=4,
        color=z1,
        colorscale='plasma',
        opacity=1
    ),
    line=dict(
        color='blue',
        width=10
    ),
    name='Pendulo 1' 
))

fig.add_trace(go.Scatter3d(
    x=x2,
    y=y2,
    z=z2,
    mode='lines+markers',
    marker=dict(
        size=4,
        color=z2,
        colorscale='plasma',
        opacity=1
    ),
    line=dict(
        color='red',  
        width=10
    ),
    name='Pendulo 2'  
))

fig.add_trace(go.Scatter3d(
    x=x3,
    y=y3,
    z=z3,
    mode='lines+markers',
    marker=dict(
        size=4,
        color=z3,
        colorscale='plasma',
        opacity=1
    ),
    line=dict(
        color='green',  
        width=10
    ),
    name='Pendulo 3'  
))

fig.update_layout(
    title='Pndulo 3D com 3 Sries de Dados',
    scene=dict(
        xaxis_title='Eixo X',
        yaxis_title='Eixo Y',
        zaxis_title='Eixo Z'
    )
)

fig.show()
```
